
Выше анализ датасета подтверждает, что данные разделены на три независимых блока, каждый из которых измеряется по-своему.

1. Блок «Отказ от курения» (Cessation) — Строки 0-1
Здесь нет вариантов ответа (Response = NaN), потому что вопрос в колонке MeasureDesc уже содержит в себе ответ.
Строка 0: Статистика по тем, кто хочет бросить.
Строка 1: Статистика по тем, кто уже пытался бросить за последний год.

2. Блок «Курение сигарет» (Cigarette Use) — Строки 2-4
Здесь используется показатель «Статус курения» (Smoking Status), который расшифровывается через колонку Response:
Current, Ever, Frequent

3. Блок «Бездымный табак» (Smokeless Tobacco) — Строки 5-7


по этой причине дальше будет изменена структура: MeasureDesc строго определяет TopicDesc. TopicDesc удалить. one-hot кодирование для MeasureDesc и ordinal encoding для response.

```python
'''

one-hot encoding

'''

cols_to_encode = ['DataSource', 'Gender', 'Education', 'MeasureDesc']
df_encoded = pd.get_dummies(df, columns=cols_to_encode, drop_first=False, dtype=int)


''' 

TopicDesc drop


'''

df_encoded = df_encoded.drop(columns='TopicDesc')



'''

Ordinal encoding

'''

response_map = {
    'Not Applicable': 0,
    'Ever': 1,
    'Current': 2,
    'Frequent': 3
}

df_encoded['Response_Ordinal'] = df_encoded['Response'].map(response_map)
df_encoded = df_encoded.drop(columns='Response')
```


LocationAbbr	Target Encoding заменит название штата на средний процент курения в нем

```python
'''

target encoding for LocationAbbr

'''

import category_encoders as ce

encoder = ce.TargetEncoder(cols=['LocationAbbr'])

df_encoded['Location_Encoded'] = encoder.fit_transform(df_encoded['LocationAbbr'], df_encoded['Data_Value'])
#df_encoded = df_encoded.drop(columns='LocationAbbr')
```

#### Добавление новых фичей:

1. Оценка абсолютного количества (Estimated_Affected_Count) 
Этот признак переводит проценты в примерное количество реальных людей, что помогает модели учитывать масштаб выборки.

2. Индикатор шумности данных (Value_Error_Interaction)
Перемножение значения на его стандартную ошибку помогает выявить записи, где высокий показатель сочетается с высокой неопределенностью.

```python
df_encoded['Estimated_Affected_Count'] = (df_encoded['Data_Value'] / 100) * df_encoded['Sample_Size']

df_encoded['Value_Error_Interaction'] = df_encoded['Data_Value'] * df_encoded['Data_Value_Std_Err']
```

## **Визуализация**

1. Распределение признаков

```python
categorical_cols =['Gender', 'Education', 'Response']

# 1. Получаем список всех уникальных тем (исключая пустые значения, если они есть)
topics = df['TopicDesc'].dropna().unique()
n_rows = len(topics) # Количество строк (равно количеству тем)
n_cols = len(categorical_cols) # Количество колонок (у нас их 3)

# 2. Создаем сетку графиков. 
# Высоту делаем динамической: 5 дюймов на каждую строку тем
fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, 5 * n_rows))

# 3. Запускаем двойной цикл: сначала по темам (строкам), затем по признакам (колонкам)
for r, topic in enumerate(topics):
    
    # Отфильтруем датафрейм, оставив только данные для текущей темы
    df_topic = df[df['TopicDesc'] == topic]
    
    for c, col in enumerate(categorical_cols):
        # Получаем нужный график из сетки (axes теперь двумерный массив)
        ax = axes[r, c] 
        
        # Считаем значения ТОЛЬКО для отфильтрованного df_topic
        counts = df_topic[col].value_counts()
        
        # Строим столбцы
        ax.bar(counts.index, counts.values, color='skyblue', edgecolor='black')
        
        # Настраиваем внешний вид. Добавляем название темы в заголовок!
        ax.set_title(f'Тема: {topic}\nБаланс: {col}', fontsize=12)
        ax.set_ylabel('Количество записей')
        ax.tick_params(axis='x', rotation=45) 
        
        # Добавляем значения над столбцами
        for index, value in enumerate(counts.values):
            ax.text(index, value, str(value), ha='center', va='bottom', fontsize=9)

plt.tight_layout() # Автоматически выравнивает графики
plt.show()
```
